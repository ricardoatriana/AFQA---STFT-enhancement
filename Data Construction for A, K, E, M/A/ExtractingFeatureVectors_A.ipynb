```python
import os
import pickle
import cv2
from afqa_toolbox.quality import ClassicEnsemble

def extract_features(image_path, output_dir, ensemble_models_path, mindet_lib_path, pca_coeffs_path, precalculated_edges_path):
    image = cv2.imread(image_path, 0)
    if image is None:
        print(f"Error: Could not load {image_path}")
        return

    filename = os.path.basename(image_path).split(".")[0]
    image500 = cv2.resize(image, None, fx=500/500, fy=500/500, interpolation=cv2.INTER_NEAREST)

    q = ClassicEnsemble(
        ensemble_models_path="/home/ricardoatriana/ensemble_models/models_randomforest.pkl",
        mindet_lib_path="/home/ricardoatriana/FingerJetFXOSE/FingerJetFXOSE/dist/FingerJetFXOSE/build/libFJFX.so",
        pca_coeffs_path="/home/ricardoatriana/OpenAFQA/afqa_toolbox/resources/pca_fusion_model.pkl",
        bin_edges_path="/home/ricardoatriana/OpenAFQA/afqa_toolbox/resources/precalculated_edges.pkl"
    )

    feature_vector = q.feature_vector(image500)

    output_file = os.path.join(output_dir, f"{filename}.pkl")
    with open(output_file, "wb") as f:
        pickle.dump(feature_vector, f)

    print(f"Saved feature vector: {output_file}")

# Extract Gabor-enhanced features
gabor_folder = "/home/ricardoatriana/GaborEnhancedImages"
output_gabor = "/home/ricardoatriana/feature_vectors_gabor/"
os.makedirs(output_gabor, exist_ok=True)

for img_file in os.listdir(gabor_folder):
    if img_file.endswith(".png"):
        extract_features(os.path.join(gabor_folder, img_file), output_gabor, "path_to_ensemble_model", "path_to_mindtct", "path_to_pca", "path_to_edges")
```
