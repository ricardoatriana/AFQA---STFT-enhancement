```python
import cv2
import os
import numpy as np
from afqa_toolbox.enhancement import GaborEnhancement, dog_filter, magnitude_filter
from afqa_toolbox.tools import normed

# Input folder containing fingerprint images
latent_folder = "/home/ricardoatriana/sd301a/dryrun-A/500/roll/png"  # Update with your folder path

# Output folder to save processed images
output_folder = "/home/ricardoatriana/GaborEnhancedImages"
os.makedirs(output_folder, exist_ok=True)  # Create output folder if it doesn't exist

# List all PNG files in the folder
image_files = [f for f in os.listdir(latent_folder) if f.endswith(".png")]

# Set PPI manually for fingerprint images
ppi = 500  

# Initialize enhancement object
enh = GaborEnhancement()

for img_file in image_files:
    img_path = os.path.join(latent_folder, img_file)
    image = cv2.imread(img_path, 0)

    if image is None:
        print(f"Error loading {img_path}")
        continue

    # Resize image to 500 PPI
    image = cv2.resize(image, None, fx=500 / ppi, fy=500 / ppi, interpolation=cv2.INTER_NEAREST)

    # Apply Gabor enhancement
    enhanced_gabor, binim = enh.enhance(image)

    # Apply DoG filter with default parameters
    enhanced_dog = dog_filter(image, ksize=5, sigma=1.0, log_enhance=True)

    # Apply magnitude filter with a factor of 0.5
    enhanced_mag, _ = magnitude_filter(image, 0.5)

    # Save processed images
    base_name = os.path.splitext(img_file)[0]  # Remove extension
    
    cv2.imwrite(os.path.join(output_folder, f"{base_name}_binary.png"), binim.astype(np.uint8) * 255)


    print(f"Processed and saved: {img_file}")

print("Processing complete. All images saved to:", output_folder)

```
