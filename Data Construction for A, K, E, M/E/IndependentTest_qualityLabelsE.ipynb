```python
import os
import pickle
import time
from afqa_toolbox.quality import ClassicEnsemble

# Define paths
image_folder = "/home/ricardoatriana/IndependentTest_toAK/E/GabEnhE"  # Folder containing 240 images
pkl_folder = "/home/ricardoatriana/IndependentTest_toAK/E/IndependentTest_featureVectors"  # Folder containing 240 .pkl files
output_txt = "/home/ricardoatriana/IndependentTest_toAK/E/IndependetTest_fused_quality_scores.txt"  # Output text file for fused quality scores

# Initialize ClassicEnsemble
q = ClassicEnsemble(
    ensemble_models_path="/home/ricardoatriana/ensemble_models/models_randomforest.pkl",
    mindet_lib_path="/home/ricardoatriana/FingerJetFXOSE/FingerJetFXOSE/dist/FingerJetFXOSE/build/libFJFX.so", #Its not needed as feature vector is already computed
    pca_coeffs_path="/home/ricardoatriana/OpenAFQA/afqa_toolbox/resources/pca_fusion_model.pkl",
    bin_edges_path="/home/ricardoatriana/OpenAFQA/afqa_toolbox/resources/precalculated_edges.pkl"
)

# Open the output file to store fused quality scores
with open(output_txt, "w") as txt_file:
    txt_file.write("Filename,FusedQuality\n")  # Write header

    # Process each image
    for filename in os.listdir(image_folder):
        if filename.endswith(".png"):  # Process only PNG images
            image_path = os.path.join(image_folder, filename)
            pkl_path = os.path.join(pkl_folder, filename.replace(".png", ".pkl"))

            # Load the feature vector from the pkl file
            with open(pkl_path, "rb") as f:
                feature_vector = pickle.load(f)

            # Compute ensemble prediction and fused quality
            start_time = time.time()
            ensemble_prediction = q.predict_ensemble(feature_vector)
            fused_quality = q.fusion(ensemble_prediction)
            elapsed_time = time.time() - start_time

            # Print and save results
            print(f"{filename} → Fused Quality: {fused_quality:.2f} (Time: {elapsed_time:.2f}s)")
            txt_file.write(f"{filename},{fused_quality:.2f}\n")  # Save to file

print(f"All results saved in {output_txt}")
```
