```python
import os
import pickle
import numpy as np

# Define file paths
feature_dir = "/home/ricardoatriana/IndependentTest_toAK/E/IndependentTest_featureVectors"  # Folder where .pkl files are stored
quality_file = "/home/ricardoatriana/IndependentTest_toAK/E/IndependetTest_fused_quality_scores.txt"  # File with fused quality scores

# Load quality scores from .txt file
quality_dict = {}
with open(quality_file, "r") as f:
    for line in f:
        parts = line.strip().split(",")  # Ensure correct delimiter
        if len(parts) == 2:
            filename = os.path.splitext(parts[0])[0]  # Remove extension
            try:
                quality = float(parts[1])
                quality_dict[filename] = quality
            except ValueError:
                print(f"❌ Invalid quality score in line: {line}. Skipping...")

# Load feature vectors and labels
X = []
y = []

for pkl_file in os.listdir(feature_dir):
    if pkl_file.endswith(".pkl"):
        base_name = os.path.splitext(pkl_file)[0]
        if base_name in quality_dict:
            try:
                # Load feature vector
                with open(os.path.join(feature_dir, pkl_file), "rb") as f:
                    feature_vector = pickle.load(f)
                
                # Ensure feature vector is a NumPy array
                feature_vector = np.array(feature_vector)

                X.append(feature_vector)
                y.append(quality_dict[base_name])
            except Exception as e:
                print(f"❌ Error loading {pkl_file}: {e}")

# Convert lists to NumPy arrays
X = np.array(X, dtype=object)  # Use `dtype=object` to handle variable-sized data
y = np.array(y)

# Save dataset
np.save("X.npy", X)
np.save("y.npy", y)

# Print dataset info
print(f"✅ Dataset saved! Ready for training.")
print(f"📊 X shape: {X.shape}, y shape: {y.shape}")
```
